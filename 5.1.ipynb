```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('http://www.cwb.gov.tw/V7/service/notice/download/notice_20171114100441.csv')
df
```

```python
df = df[df['規模'] != '1≦M<2']
df
```

```python
df = df[df['規模'] != '2≦M<3']
df
```

```python
df.drop(['規模'], axis=1)
```

## 年份與地震規模大於3的關係

```python
y = []
eqnum = []

for i in range(1994, 2017):
    y.append(i)
    eqnum.append(df[df['年份']==i]['地震個數'].sum())
```

```python
data={"year" : y, "eqnum" : eqnum}
df=pd.DataFrame(data)
df
```

```python
plt.scatter(df.year,df.eqnum)
```

```python
from sklearn.model_selection import train_test_split
```

```python
df.year.values
```

```python
X = df.year.values.reshape(len(df.year.values), 1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, df.eqnum.values, test_size=0.3, random_state = 9487)
```

```python
plt.scatter(x_train, y_train)
```

```python
plt.scatter(x_test, y_test)
```

```python
regr = LinearRegression()
```

```python
regr.fit(x_train, y_train)
```

```python
Y_pred = regr.predict(x_test)
```

```python
plt.scatter(x_test, y_test)
plt.plot(x_test, Y_pred, 'r')
```

```python
plt.scatter(X, df.eqnum.values)
plt.plot(x_test, Y_pred, 'r')
```

## 規模大於3的地震數量近幾年有逐年攀升的趨勢
