# 近年寵物遺失通報案件

近期在facebook上看到很多寵物走失案件，因而好奇查看政府統計寵物遺失案件

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
```

```python
df = pd.read_csv('http://data.coa.gov.tw/Service/OpenData/TransService.aspx?UnitId=IFJomqVzyB0i&FOTT=CSV')
df
```

```python
dfcat = df[df['寵物別']=='貓']
dfdog = df[df['寵物別']=='狗']
```

```python
dfcatB = dfcat[dfcat['性別']=='公']
dfcatG = dfcat[dfcat['性別']=='母']
dfdogB = dfdog[dfdog['性別']=='公']
dfdogG = dfdog[dfdog['性別']=='母']
```

```python
cat = [len(dfcatB), len(dfcatG)]
dog = [len(dfdogB), len(dfdogG)]
```

```python
myfont = FontProperties(fname=r'C:\Users\julia\Desktop\font\NotoSansCJKtc-Medium.otf')
```

```python
sex = ['Boy','Girl']
```

```python
plt.title('貓狗公母走失通報數量',fontproperties=myfont)
plt.xlabel('性別',fontproperties=myfont)
plt.ylabel('數量',fontproperties=myfont)
plt.bar(sex, cat, label = 'cat', align = 'edge', width = 0.35, color = 'violet')
plt.bar(sex, dog, label = 'dog', align = 'edge', width = -0.35, color = 'royalblue')
plt.legend(facecolor='lemonchiffon')
```
