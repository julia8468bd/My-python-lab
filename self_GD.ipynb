# 自訂義Gradient Descent

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#自己找函數跟範圍
x = np.linspace(-10, 10, 1000)
y = (x-5)*(x+4)*(x-8)*(x+7)
```

```python
#先畫出來看看是不是好函數(有無lacal minimum)
plt.plot(x, y)
```

```python
#定義函數還有它的微分
def function(x):
    return (x-5)*(x+4)*(x-8)*(x+7)

#手動微分一下
def dfunction(x):
    return 4*x**3-6*x**2-150*x+76
```

```python
#最重要的Gradient_Descent函式
#可調參數有：起始點、學習率(learning rate)、疊代次數

def Gradient_Descent(start_point, lr, epochs):
    point = start_point
    
    plt.plot(x, function(x), c='g')
    for i in range(epochs):
        point = point - lr*dfunction(point)
        print(point)
        plt.scatter(point, function(point), c='b')
    plt.scatter(point, function(point), c='r')
```

# 以下比較不同learning rate會造成不同的結果

### learning rate 太大會造成學習不穩定(常會有走過頭的情況)

```python
test1 = Gradient_Descent(0,0.01,10)
```

### learning rate 適中學習效率好，順利到達local minimum

```python
test2 = Gradient_Descent(0,0.005,10)
```

### learning rate 過小學習效率慢，難以到達local minimum

```python
test3 = Gradient_Descent(0,0.001,10)
```
