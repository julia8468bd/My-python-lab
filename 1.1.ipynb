```python
import numpy as np
import matplotlib.pyplot as plt

p21 = [-7, -7]
p22 = [3, 1.5]
p31 = [-8, -6]
p32 = [2, 2]
p41 = [-7, -8]
p42 = [1.5, 0]
p51 = [-7, -6]
p52 = [1.5, 0]

hc = np.linspace(0, np.pi, 10)
c = np.linspace(0, 2*np.pi, 500)
show1 = np.arange(-10, 10, 0.5)
show2 = np.random.randint(0 ,10, 40)

sh1 = 10*np.cos(hc)
sh2 = 10*np.sin(hc)
door1 = 4*np.cos(hc)
door2 = 4*np.sin(hc)
p11 = 1*np.cos(c)-7
p12 = 1*np.sin(c)+4

plt.xkcd()
plt.plot(sh1, sh2, 'b')
plt.plot(door1, door2, 'b')
plt.plot(show1, show2, '.')
ax = plt.gca()
ax.set_aspect('equal')
plt.plot(p11, p12, 'lightsalmon')
plt.plot(p21, p22, 'lightsalmon')
plt.plot(p31, p32, 'lightsalmon')
plt.plot(p41, p42, 'lightsalmon')
plt.plot(p51, p52, 'lightsalmon')
plt.title('Snow House')
```
